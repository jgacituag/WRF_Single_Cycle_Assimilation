```python
import os
import glob
import numpy as np
import pandas as pd

# 1. Define where your cluster dumped the files
outdir = "../data/ws1_below_test/"  # Update this to your actual output path
tag = "EXP"                         # Update this to match your experiment_tag

# 2. Find all the chunk files
search_pattern = os.path.join(outdir, f"{tag}_fullgrid_chunk*_True*.npy")
chunk_files = glob.glob(search_pattern)
chunk_files.sort()

print(f"Found {len(chunk_files)} chunk files to process.")

# 3. Load and concatenate everything
all_data = []
for f in chunk_files:
    # Load the list of dicts from the .npy file
    chunk_data = np.load(f, allow_pickle=True)
    all_data.extend(chunk_data)

# 4. Convert the massive list into a single Pandas DataFrame
df = pd.DataFrame(all_data)

print(f"Successfully combined {len(df)} assimilated grid points!")
display(df.head())

# 5. Save the combined DataFrame for instant loading next time
# Pickle preserves the internal NumPy arrays perfectly
combined_path = os.path.join(outdir, f"{tag}_fullgrid_COMBINED.pkl")
df.to_pickle(combined_path)

print(f"Saved master DataFrame to: {combined_path}")
```
